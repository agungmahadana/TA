# *CLEANING*
Pada tahapan ini bertujuan untuk membershikan data dari atribut yang tidak diperlukan.

```python
import re

def clean_text(df, text_field, new_text_field_name):
    # Menyalin nilai kolom text_field ke kolom baru, mengonversi ke string terlebih dahulu
    df[new_text_field_name] = df[text_field].astype(str)

    # Mengubah seluruh teks menjadi lowercase
    df[new_text_field_name] = df[new_text_field_name].str.lower()

    # Menghapus hashtag (#) dan mentions (@)
    df[new_text_field_name] = df[new_text_field_name].apply(lambda elem: re.sub(r'(@\w+|#\w+)', ' ', elem))

    # Menghapus URL atau link (http://, https://, dan lainnya)
    df[new_text_field_name] = df[new_text_field_name].apply(lambda elem: re.sub(r'http\S+|www\S+', ' ', elem))

    # Menghapus huruf berlebihan (lebih dari 2 kali berturut-turut)
    df[new_text_field_name] = df[new_text_field_name].apply(lambda elem: re.sub(r'(\w)\1{2,}', r'\1', elem))

    # Menghapus angka, karakter non-alfanumerik, dan karakter non-ASCII dalam satu langkah
    df[new_text_field_name] = df[new_text_field_name].apply(lambda elem: re.sub(r'[^a-zA-Z\s]', ' ', elem))

    # Menghapus newline
    df[new_text_field_name] = df[new_text_field_name].apply(lambda elem: elem.replace('\n', ' '))

    # Menghapus leading/trailing whitespace
    df[new_text_field_name] = df[new_text_field_name].apply(lambda elem: elem.strip())

    # Mengganti spasi berlebih menjadi satu spasi
    df[new_text_field_name] = df[new_text_field_name].apply(lambda elem: re.sub(r'\s+', ' ', elem))

    return df

# Menerapkan case folding
data_clean = clean_text(combined_df, 'content', 'text_clean')

# Menghapus baris yang kosong dan menghapus duplikat
data_clean = data_clean[data_clean['text_clean'] != '']
data_clean = data_clean.drop_duplicates(subset=['text_clean'])

data_clean
```

# *TOKENIZING*
Tokenizing adalah proses pemisahan teks menjadi potongan-potongan yang disebut sebagai token untuk kemudian di analisa. Kata, angka, simbol, tanda baca dan entitas penting lainnya dapat dianggap sebagai token. Didalam NLP, token diartikan sebagai ‚Äúkata‚Äù meskipun tokenize juga dapat dilakukan pada paragraf maupun kalimat.

```python
from nltk.tokenize import word_tokenize

# Menerapkan tokenizing
data_clean['text_tokens'] = data_clean['text_clean'].apply(lambda x: word_tokenize(x))

data_clean
```

# *FORMALIZING*
Formalizing adalah proses pengantian kata-kata slang ke bentuk formalnya.

```python
# Fungsi untuk mengganti slang dengan bentuk formal
def replace_slang_simple(tokens, kamusSlang):
    # Membuat list untuk token yang sudah diganti
    new_tokens = []
    changed_words = []  # Menyimpan kata yang diganti
    
    for kata in tokens:
        # Jika kata ada dalam kamus slang, ganti dengan bentuk formal
        if kata in kamusSlang:
            new_tokens.append(kamusSlang[kata])
            changed_words.append((kata, kamusSlang[kata]))
        else:
            new_tokens.append(kata)  # Jika tidak ada, tetap menggunakan kata asli
    
    return new_tokens, changed_words

# Membaca kamus slang dari file .xlsx menggunakan pandas
def load_slang_file_from_xlsx(file_path):
    # Membaca file Excel
    df = pd.read_excel(file_path)
    # Mengonversi menjadi kamus (Slang -> Formal)
    return dict(zip(df['Slang'], df['Formal']))

# Membaca kamus slang dari file .xlsx
kamusSlang = load_slang_file_from_xlsx('slangwords.xlsx')

# Menambahkan kolom untuk kata yang diganti dan jumlah perubahan
changed_words_list = []

# Proses penggantian kata slang
def replace_slang(row):
    # Asumsi bahwa data yang dimasukkan sudah berupa list token
    new_tokens, changed_words = replace_slang_simple(row, kamusSlang)
    changed_words_list.extend(changed_words)  # Menyimpan kata yang diganti
    return new_tokens  # Mengembalikan list token yang sudah diganti

# Menerapkan penggantian kata slang pada kolom 'text_tokens' (sudah tokenized)
data_clean['text_formal'] = data_clean['text_tokens'].apply(lambda row: replace_slang(row))

# Menghapus baris kosong dan duplikat
data_clean = data_clean[data_clean['text_formal'] != '']
data_clean = data_clean.drop_duplicates(subset=['text_clean'])

# Menghitung jumlah total kata yang diganti
print(f"Jumlah total kata yang diganti: {len(changed_words_list)}")

data_clean
```

# *STOPWORD REMOVAL*
Stopword adalah kata umum yang biasanya muncul dalam jumlah besar dan dianggap tidak memiliki makna. Contoh stopword dalam bahasa Indonesia adalah emoticon ‚Äúyang‚Äù, ‚Äúdan‚Äù, ‚Äúdi‚Äù, ‚Äúdari‚Äù, dll. Makna di balik penggunaan stopword yaitu dengan menghapus kata-kata yang memiliki informasi rendah dari sebuah teks, kita dapat fokus pada kata-kata penting sebagai gantinya.

```python
import nltk
from nltk.corpus import stopwords
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

# Mengambil daftar stopword dari NLTK (Bahasa Indonesia dan Inggris)
stop_nltk_indonesia = set(stopwords.words('indonesian'))
stop_nltk_english = set(stopwords.words('english'))

# Mengambil daftar stopword dari Sastrawi
factory = StopWordRemoverFactory()
stop_sastrawi = set(factory.get_stop_words())

# Menambahkan kata ke dalam daftar stopword
custom_stopwords = {'yg', 'nya'}

# Menggabungkan stopword dari NLTK, Sastrawi, dan custom stopwords
combined_stopwords = stop_nltk_indonesia | stop_nltk_english | stop_sastrawi | custom_stopwords

# Fungsi untuk menghapus stopword dari list token
def remove_stopwords(tokens):
    # Menghapus stopword dari list token
    return [word for word in tokens if word not in combined_stopwords]

# Menerapkan penghapusan stopword pada kolom 'text_tokens' (sudah tokenized)
data_clean['text_stopWord'] = data_clean['text_formal'].apply(remove_stopwords)

# Menghapus baris yang kosong dan menghapus duplikat
data_clean = data_clean[data_clean['text_stopWord'] != '']
data_clean = data_clean.drop_duplicates(subset=['text_stopWord'])

data_clean
```

# *STEMMING*
Stemming adalah proses pemetaan dan penguraian bentuk dari suatu kata menjadi bentuk kata dasarnya. Untuk melakukan stemming bahasa Indonesia kita dapat menggunakan library Python Sastrawi yang sudah kita siapkan di awal. Library Sastrawi menerapkan Algoritma Nazief dan Adriani dalam melakukan stemming bahasa Indonesia.

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Membuat stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Fungsi untuk melakukan stemming pada suatu term
def stemmed_wrapper(term):
    # Pastikan term adalah string sebelum diproses
    return stemmer.stem(str(term))

# Fungsi untuk melakukan stemming pada setiap token dalam list (document)
def stem_tokens(tokens):
    return [stemmed_wrapper(term) for term in tokens]

# Menyimpan hasil stemming dalam dictionary dan menghitung jumlah kata yang diubah
term_dict = {}
hitung = 0

# Menerapkan stemming pada kolom 'text_stopWord' dan menyimpan hasilnya dalam kolom baru 'text_steamIndo'
def process_stemming(row):
    global hitung
    # Lakukan stemming pada setiap token dalam row
    stemmed_tokens = stem_tokens(row)
    
    # Menyimpan hasil stemming dan menghitung perubahan kata
    for original, stemmed in zip(row, stemmed_tokens):
        if original != stemmed and original not in term_dict:
            term_dict[original] = stemmed
            hitung += 1
            print(f"{hitung}: {original} -> {stemmed}")
    
    return stemmed_tokens

# Menerapkan stemming pada kolom 'text_stopWord' (setelah stopword removal)
data_clean['text_steamIndo'] = data_clean['text_stopWord'].apply(process_stemming)

# Menghapus baris yang kosong dalam kolom 'text_steamindo' dan menghapus duplikat
data_clean = data_clean[data_clean['text_steamIndo'] != '']
data_clean = data_clean.drop_duplicates(subset=['text_steamIndo'])

# Menampilkan hasil
print('------------------------------')
print(f"Hasil Stemming {hitung} Kata:")
print(term_dict)
print('------------------------------')

data_clean
```

# **3. MODELING**
Memecah data test 20% dari keseluruhan data membagi data Anda menjadi set pelatihan dan set pengujian. Set pelatihan digunakan untuk melatih model Naive Bayes, sedangkan set pengujian digunakan untuk menguji kinerja model.

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(df_undersampled['text_steamIndo'], df_undersampled['label'], test_size = 0.2, random_state = 42)

# Pastikan semua nilai NaN diisi atau dihapus, dan semua elemen dikonversi ke string
X_train = X_train.fillna("").astype(str)
X_test = X_test.fillna("").astype(str)

# Vektorisasi dengan TF
tf_vectorizer = CountVectorizer()
tf_train = tf_vectorizer.fit_transform(X_train)
tf_test = tf_vectorizer.transform(X_test)

# Melatih model dan prediksi
nb = MultinomialNB()
nb.fit(tf_train, y_train)
y_pred = nb.predict(tf_test)

# Evaluasi
print("MultinomialNB Accuracy:", accuracy_score(y_test, y_pred))
print("MultinomialNB Precision (macro):", precision_score(y_test, y_pred, average="macro"))
print("MultinomialNB Recall (macro):", recall_score(y_test, y_pred, average="macro"))
print("MultinomialNB F1 Score (macro):", f1_score(y_test, y_pred, average="macro"))
print(f'Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}')
print('\n=====================================================\n')
print(classification_report(y_test, y_pred, zero_division=0))
```
